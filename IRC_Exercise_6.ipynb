```python
# Import the necessary libraries 
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

%matplotlib inline
```

```python
# Load data files 
base_path = '/Users/Manu/Desktop/Introduction-to-Research-Computing-master/Examples/Data'

filename_1 = 'Period1.txt'
filename_2 = 'Period2.txt'
filename_3 = 'Period3.txt'
filename_4 = 'Period4.txt'

path_to_file_1 = os.path.join(base_path, filename_1)
path_to_file_2 = os.path.join(base_path, filename_2)
path_to_file_3 = os.path.join(base_path, filename_3)
path_to_file_4 = os.path.join(base_path, filename_4)

data_1 = np.loadtxt(path_to_file_1)
data_2 = np.loadtxt(path_to_file_2)
data_3 = np.loadtxt(path_to_file_3)
data_4 = np.loadtxt(path_to_file_4)
```

```python
# Plot the first data set 
plt.plot(data_1)
```

```python
# Find the peaks and check if
# you are finding them properly
peaks, _ = find_peaks(data_1, height = 2.0, distance = 200)

plt.plot(data_1)
plt.plot(peaks, data_1[peaks], "x")
```

```python
# Choose the width of the peaks 
# and plot them 
range_peak = 100
fig = plt.figure(figsize=(3,6))
plt.grid()

for i in range(len(peaks)):
    plt.plot(data_1[peaks[i] - range_peak : peaks[i] + 2*range_peak])
```

```python
# Now the second data set 
plt.plot(data_2)
```

```python
# In this case it is better to invert
# the dataset and work with the minima instead
data_2_test = np.copy(data_2)
data_2_test = -1.0* data_2_test

peaks_2, _ = find_peaks(data_2_test, height = -2.4, distance = 140)

plt.plot(data_2_test)
plt.plot(peaks_2, data_2_test[peaks_2], "x")
```

```python
# Choose the width of the maxima 
range_peak_2 = 50
fig2 = plt.figure(figsize=(3,6))

for i in range(len(peaks_2)):
    plt.plot(data_2[peaks_2[i] - int(2*range_peak_2) : peaks_2[i] + range_peak_2])
    plt.grid()
```
